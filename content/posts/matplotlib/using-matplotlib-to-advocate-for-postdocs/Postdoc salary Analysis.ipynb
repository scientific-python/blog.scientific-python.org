```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
salary = {2016: [43692, 45444, 47268, 49152, 51120, 53160, 55296, 57504],
         2017: [47484, 47844, 48216, 50316, 52140, 54228, 56400, 58560],
         2018: [48432, 48804, 49188, 51324, 53184, 55308, 57528, 59736],
         2019: [50004, 50376, 50760, 52896, 54756, 56880, 59100, 61308]}
inflation = {2017: 2.1,
            2018: 1.9,
            2019: 1.7}
housing = {2016: 2963,
           2017: 3263,
           2018: 3298,
           2019: 3505}
```

```python
plt.figure(figsize=(6, 4))
colors = plt.cm.Reds_r(np.linspace(0.6, 0, len(salary.keys())))
for i, year in enumerate(salary.keys()):
    plt.plot(np.array(salary[year]), "o-", label=year, color=colors[i])
for i, year in enumerate(salary.keys()):
    if year == 2019:
        continue
    plt.plot(np.array(salary[year])*(1+inflation[year+1]/100), "--", color=colors[i+1], label="%d+inflation" % year)
plt.xticks(range(8), list(range(8)))
plt.xlabel("Years of postdoc experience")
plt.ylabel("Gross annual salary (USD)")
plt.text(-1.4, 40200, "source: NIH", fontsize=8)
plt.legend(ncol=2)
plt.savefig("gross_salary.png", dpi=200)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
colors = plt.cm.Reds_r(np.linspace(0.6, 0, len(salary.keys())))
for i, year in enumerate(salary.keys()):
    plt.plot(np.array(salary[year])-housing[year]*12, "o-", label=year, color=colors[i])
plt.xticks(range(8), list(range(8)))
plt.xlabel("Years of postdoc experience")
plt.ylabel("Gross annual salary minus average rent (USD)")
plt.text(-1.4, 5300, "source: NIH and rentcafe.com", fontsize=8)
plt.legend()
plt.savefig("gross_salary_minus_rent.png", dpi=200)
plt.show()
```
